```python
import numpy as np
import calendar
import seaborn as sns
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
import ast 
df = pd.read_csv(r"C:\Users\homeA\Desktop\Python - Youtube\data_jobs.csv")

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)


```

```python
# df = df.dropna(subset='salary_year_avg') 
df_us = df[(df['job_country']=='United States') & (df['job_title']=='Data Analyst')].copy()

df_us['job_posted_month_num'] = df_us['job_posted_date'].dt.month
```

```python
df_us_explode = df_us.explode('job_skills')
```

```python
df_us_pivot=df_us_explode.pivot_table(index = 'job_posted_month_num', columns='job_skills', aggfunc='size', fill_value=0)

df_us_pivot.loc['Total'] = df_us_pivot.sum()

df_us_pivot=df_us_pivot[df_us_pivot.loc['Total'].sort_values(ascending=False).index]

df_us_pivot = df_us_pivot.drop('Total')
```

```python
da_total = df_us.groupby('job_posted_month_num').size()

df_us_percent = df_us_pivot.div(da_total/100, axis=0)
```

```python
df_us_percent = df_us_percent.reset_index()
df_us_percent['job_posted_month'] = df_us_percent['job_posted_month_num'].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%b'))
df_us_percent
df_us_percent = df_us_percent.set_index('job_posted_month')
df_us_percent = df_us_percent.drop(columns='job_posted_month_num')
df_us_percent
```

```python
df_plot = df_us_percent.iloc[:,:5]
sns.lineplot(data=df_plot, dashes=False, palette='tab10')
sns.despine()
sns.set_theme(style='ticks')
plt.xlabel('2023')
plt.legend().remove()
plt.ylabel('Likelihood in job posting')
plt.title('Trending Top Skills for Data Analysis in The US')
from matplotlib.ticker import PercentFormatter
ax = plt.gca()
ax.yaxis.set_major_formatter(PercentFormatter(decimals=0))
for i in range(5):
    plt.text(11.2, df_plot.iloc[-1,i], df_plot.columns[i])
```
