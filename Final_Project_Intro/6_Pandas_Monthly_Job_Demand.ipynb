```python
## ChatGPT Explaination


# Step 1: Create a copy of the DataFrame containing only jobs in the United States.
df_us = df[df['job_country'] == 'United States'].copy()

# Step 2: Create a new column 'job_posted_month' to store the month names (e.g., "January", "February", etc.).
df_us['job_posted_month'] = df_us['job_posted_date'].dt.strftime('%B')

# Step 3: Create a pivot table to count the number of job postings per month for each job title.
# - 'index' → Groups by 'job_posted_month' (row-wise).
# - 'columns' → Creates separate columns for each unique 'job_title_short'.
# - 'aggfunc="size"' → Counts the number of job postings for each job title.
df_us_pivot = df_us.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')
df_us_pivot

# Step 4: Reset the index so that 'job_posted_month' becomes a regular column instead of an index.
df_us_pivot = df_us_pivot.reset_index()
df_us_pivot

# Step 5: Create a new column 'month_number' to store the numerical month representation.
# - This is necessary for sorting the months in chronological order.
# - Converts month names (e.g., "January") into numbers (1 for January, 2 for February, etc.).
df_us_pivot['month_number'] = pd.to_datetime(df_us_pivot['job_posted_month'], format='%B').dt.month
df_us_pivot


# Step 6: Sort the DataFrame based on 'month_number' to arrange months in chronological order.
df_us_pivot.sort_values('month_number', inplace=True)


# Step 7: Set 'job_posted_month' back as the index so that the table is structured correctly.
# - After sorting, months now appear in proper order from January to December.
df_us_pivot.set_index('job_posted_month', inplace=True)
df_us_pivot



# Step 8: Remove the 'month_number' column as it's no longer needed after sorting.
df_us_pivot.drop(columns='month_number', inplace=True)



# Step 9: Identify the **top 3 job titles** with the highest number of postings.
# - `value_counts()` counts occurrences of each job title.
# - `head(3)` selects the top 3 most frequent job titles.
top_3 = df_us['job_title_short'].value_counts().head(3)



# Step 10: Convert the index (which contains job titles) into a list.
# - This list will be used to filter job titles for visualization.
top_3 = top_3.index.tolist()



# Step 11: Plot the trends of the **top 3 job titles** over the months.
# - This generates a **line plot** to visualize job posting trends over time.
df_us_pivot[top_3].plot(kind='line')
```
