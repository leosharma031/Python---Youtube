```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\homeA\Desktop\Python - Youtube\data_jobs.csv")

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
## Merge US Jobs and Monthly Postings (2.7.1) - Problem

# Problem Statement:
# Create a DataFrame df_us_jobs that contains job postings in the United States.
# Then create another DataFrame df_monthly_postings that counts the number of job postings for each month.
# Merge these two DataFrames on the job_posted_month column using a right join and display the first 5 rows of the resulting DataFrame.


df_us = df[df['job_country'] == 'United States'].copy()
df_us['job_posted_month'] = df_us['job_posted_date'].dt.to_period('M')
# df_us['job_posted_month'] = df_us['job_posted_date'].dt.strftime('%B')

# df_monthly_posting = df_us.groupby('job_posted_month').size()

df_monthly_posting = df_us.groupby('job_posted_month').size().reset_index(name='number_of_jobs')
df_monthly_posting
merged_df = pd.merge(df_us, df_monthly_posting, on='job_posted_month',how='right')
merged_df.head()
```

```python
# Merge Non-Remote Jobs and Salary (2.7.2) - Problem
# Pandas Merge DataFrames
# Problem Statement:

# Create a DataFrame df_not_remote_jobs that filters job postings that do not offer remote work.
# Then create another DataFrame df_not_remote_salary that calculates the average yearly salary for 
# these non-remote jobs. Rename this salary column salary_not_remote_avg.
# Merge these two DataFrames on the job_title_short column using a right join and display the 
# first 5 rows of the resulting DataFrame.

df_not_remote_jobs = df[df['job_work_from_home']==False].copy()

df_not_remote_salary =df_not_remote_jobs.groupby('job_title_short')['salary_year_avg'].mean().reset_index()

df_not_remote_salary['salary_not_remote_avg'] = df_not_remote_salary['salary_year_avg'].drop(columns='salary_year_avg')

df_not_remote_salary=df_not_remote_salary.drop(columns= 'salary_year_avg')

pd.merge(df_not_remote_jobs, df_not_remote_salary, on ='job_title_short', how='right').head()
```

```python
# Merge Company Salary and Remote Count (2.7.3) - Problem

# Problem Statement:
# Create a DataFrame df_company_salary that calculates the average yearly salary for each company.
# Then, create another DataFrame df_company_remote that counts the number of remote job postings 
# for each company. Name this count column remote_count.
# Merge these two DataFrames on the company_name column using an inner join and display the first 5 rows 
# of the resulting DataFrame.


df_company_salary = df.copy()

df_company_salary=df_company_salary.groupby('company_name')['salary_year_avg'].mean()

df_company_remote = df[df.job_work_from_home == True].copy()

df_company_remote = df_company_remote.groupby('company_name')['job_work_from_home'].count().reset_index(name = 'remote_count')

pd.merge(df_company_salary,df_company_remote, on='company_name', how='inner').head()
```
