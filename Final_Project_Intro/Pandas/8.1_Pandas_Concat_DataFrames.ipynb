```python
import numpy as np
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# dataset = load_dataset('lukebarousse/data_jobs')
# df = dataset['train'].to_pandas()

df = pd.read_csv(r"C:\Users\homeA\Desktop\Python - Youtube\data_jobs.csv")

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
# Concat Job Postings (2.8.1) - Problem
# Pandas Concat DataFrames
# Problem Statement:
# Concatenate two DataFrames to analyze job postings for Data Scientists and Data Analysts:

# Create one DataFrame with job postings for Data Scientists named data_scientist_df and another for 
# Data Analysts named data_analyst_df.
# Concatenate them and display the combined DataFrame called combined_df.

data_scientist_jobs = pd.DataFrame({
    "Job Title": ["Data Scientist", "Machine Learning Engineer", "AI Researcher"],
    "Company": ["Company A", "Company B", "Company C"],
    "Location": ["New York", "San Francisco", "Remote"],
    "Salary": [120000, 135000, 110000]
})

data_analyst_jobs = pd.DataFrame({
    "Job Title": ["Data Analyst", "Business Intelligence Analyst", "Reporting Analyst"],
    "Company": ["Company X", "Company Y", "Company Z"],
    "Location": ["Chicago", "Austin", "Seattle"],
    "Salary": [80000, 85000, 78000]
})

combined_df = pd.concat([data_scientist_jobs, data_analyst_jobs ],ignore_index=True)
combined_df
```

```python
# Concat Monthly Postings (2.8.2) - Problem
# Pandas Concat DataFrames
# Problem Statement:
# Concatenate DataFrames to analyze job postings by month:


# Create DataFrames for job postings for the last quarter (October, November, and December).
# These are respectively called oct_df, nov_df, and dec_df.
# Concatenate them and display the combined DataFrame named combined_df.



df['job_posted_month'] = df['job_posted_date'].dt.strftime('%b')


oct_df = df[df['job_posted_month'] == 'Oct'].copy()
nov_df = df[df['job_posted_month'] == 'Nov'].copy()
dec_df = df[df['job_posted_month'] == 'Dec'].copy()

combined_df = pd.concat([oct_df, nov_df, dec_df], ignore_index=True)
combined_df.head()
```

```python
# Concat Job Types (2.8.3) - Problem
# Pandas Concat DataFrames
# Problem Statement:
# Concatenate DataFrames to analyze job postings for different job types:

# Create DataFrames for 'Full-time' roles (called full_time_df) and 'Part-time' roles (called part_time_df) job postings.
# Concatenate them and display the combined DataFrame named combined_df.



full_time_df = df[df['job_schedule_type']=='Full-time'].copy()
part_time_df = df[df['job_schedule_type']=='Part-time'].copy()
```

```python
combined_df = pd.concat([full_time_df,part_time_df], ignore_index=True)
combined_df
```
