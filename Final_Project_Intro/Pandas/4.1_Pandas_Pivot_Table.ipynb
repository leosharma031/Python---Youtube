```python
import numpy as np
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
## Count Job Postings by Title and Country (2.4.1) - Problem
##  Create a pivot table to count the number of job postings for each job_title_short and job_country. 
##  Display the first 10 rows of the resulting pivot table.

df.pivot_table(index=['job_title_short','job_country'] , aggfunc='size').head(20)
```

```python
## Salary Stats by Job Title (2.4.2) - Problem
#  Create a pivot table to find the maximum, minimum, and average salary_year_avg 
#  for each company_name. Display the first 5 rows of the resulting pivot table.

df.pivot_table(values='salary_year_avg', index = 'company_name', aggfunc=['max','min','mean']).head(5)
```

```python
## Sum Salary by Title and Country (2.4.3) - Problem
# Problem Statement:
# Create a pivot table to calculate the median of salary_year_avg for each company_name and job_country:

# Filter the pivot table to only include 'Data Scientist' roles.
# Then filter the pivot table to only include company_name with a median salary_year_avg greater than 200,000.
# Display the first 10 rows of the resulting pivot table.
# Hint:

###------------------------------------------------------------------------------------------------------------

## 1. Here, first we are passing only Data Scientist as a job title as intent to see that only
## 2. We want median of salary_year_avg, that is why it is being passed in 'values'
## 3. We want to aggregate the data by company_name and job_country, so it is being passed in 'index'
df_pivot = df[df.job_title_short == 'Data Scientist'].pivot_table(values= 'salary_year_avg', index=['company_name','job_country'],  aggfunc = 'median')

## 4. Now we want only data that has the median salary greater than of 200000
filteredtt = df_pivot[df_pivot['salary_year_avg'] > 200000]
filteredtt.head(10)
```
