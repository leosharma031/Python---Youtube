```python
import numpy as np
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
# Drop Missing Values (2.3.1) - Problem
#  Create a copy of the DataFrame and drop rows where the 
#  salary_year_avg column has missing values. Display the first 5 rows of the cleaned DataFrame.

df_copy = df.copy()
df_copy['salary_year_avg']

df_copy.dropna(subset=['salary_year_avg'], inplace=True)

df_copy.head(5)
```

```python
## Random Sample of DataFrame (2.3.2) - Problem
## Create a new DataFrame that is a random sample of 10% of the original DataFrame. Display the first 5 rows of the sampled DataFrame.

df_new = df.sample(frac=0.1)
df_new.head(5)
```

```python
# Fill Missing Salaries (2.3.3) - Problem

#  Create a copy of the original DataFrame and fill in missing values in the 
#  salary_year_avg column with the median salary. Display the first 5 rows of the altered DataFrame.




median_salary = df['salary_year_avg'].median()  
## Calculates the median salary from the 'salary_year_avg' column and stores it in the variable 'median_salary'.

median_salary  
## Displays the calculated median salary.

df_copy = df.copy()  
## Creates a copy of the original DataFrame to avoid modifying the original data.

df_copy['salary_year_avg'] = df_copy['salary_year_avg'].fillna(median_salary)  
## Replaces all missing (NaN) values in the 'salary_year_avg' column with the median salary.

df_copy.head(5)  
## Displays the first 5 rows of the modified DataFrame to check if missing values have been filled.

df_copy  
## Displays the entire modified DataFrame.
```
